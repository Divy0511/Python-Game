```python
import random

print("--- Welcome to High-Low ---")
print("Start with 100 points. Each round a card will be drawn and shown.")
print("Select whether you think the 2nd card will be Higher or Lower than the")
print("1st card.")
print("Then enter the amount you want to bet.")
print("If you are right, you win the amount you bet, otherwise you lose.")
print("Try to make it to 500 points within 10 tries.")

points = 100


def getCardValue():
    generated = random.randint(2, 14)
    return generated


def getCardStr(card_value):
    if card_value > 10:
        number_values = {"14": "A", "11": "J", "12": "K", "13": "Q"}
        return number_values.get(str(card_value), str(card_value))
    else:
        return str(card_value)


def getHLGuess():
    asking = input("Higher or Lower (H/L): ").upper()
    if asking == "H":
        return "HIGH"
    else:
        return "LOW"


def getBetAmount(maximum):
    while True:
        maximum = int(input("Input bet amount: "))
        if maximum <= points:
            return maximum
        maximum = points - maximum


def playerGuessCorrect(card1, card2, betType):
    if card1 > card2 and betType == "HIGH":
        return False
    elif card1 > card2 and betType == "LOW":
        return True
    elif card1 < card2 and betType == "HIGH":
        return True
    elif card1 < card2 and betType == "LOW":
        return False


def main():
    global points
    round_number = 1
    win_rounds = 0
    lose_rounds = 0

    while round_number <= 10 and points < 500 and points > 0:
        print("\nOVERALL POINTS: {} ROUND {}/10".format(points, round_number))
        card1 = getCardValue()
        print("First card is a [{}]".format(getCardStr(card1)))
        guess = getHLGuess()
        bet = getBetAmount(points)
        card2 = getCardValue()
        print("Second card is a [{}]".format(getCardStr(card2)))

        if playerGuessCorrect(card1, card2, guess):
            points += bet
            win_rounds += 1
        else:
            points -= bet
            lose_rounds += 1

        result = 'YOU WON' if playerGuessCorrect(card1, card2, guess) else 'YOU LOST'
        print("Card1 [{}] Card 2 [{}] - You bet '{}' for {} - {}".format(getCardStr(card1), getCardStr(card2), guess, bet, result))

        if points == 0:
            print("\n----------------LOSE--------------------")
            print("YOU RAN OUT OF POINTS AFTER {} ROUNDS!".format(round_number - 1))
            if win_rounds > 0:
                print("BUT COULD WIN {} ROUND{} AFTER {} ROUNDS!".format(win_rounds, 'S' if win_rounds > 1 else '', round_number - 1))
                print("\n----------------------------------------")
            break

        round_number += 1
        
        if round_number==11:
            print("\n---------------Lose---------------------")
            print("ONLY *{}* POINTS IN 10 ROUNDS!\nYOU COULD WIN ONLY *{}* ROUNDS OUT OF 10 ROUNDS!".format(points,win_rounds))
            print("\n----------------------------------------")

    if points >= 500:
        print("\n----------------WIN---------------------")
        print("YOU MADE IT TO {} POINTS IN {} ROUNDS!".format(points, win_rounds))
        print("-----------------------------------------")
        
main()
```
